```python
import pandas as pd
from googletrans import Translator
```

```python
%%time
df=pd.read_excel("/Users/macbookpro/Desktop/For Graduation/T3Test.xlsx")
```

```python
%%time
for row in range(0,10):
    name=df.at[row,"car_name"].strip()
    df.at[row,"car_name_t"] = translator.translate(name , dest ='en').text
    
```

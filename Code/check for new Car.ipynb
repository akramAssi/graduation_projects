```python
Cat_vars = ['make','model','type','color','fuel','history','gear','payment_method','windows']
```

```python
# we make a new DF that contains just categorial Data 
MyDataCat = MyData[Cat_vars]
MyDataCat
```

```python
MyDataCat.info()
```

```python
# in this steps we encode categorial Data 
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown='ignore')

afterEnc = enc.fit(MyDataCat)
afterEnc
```

```python
MyData.drop(Cat_vars,inplace=True,axis=1)
```

```python
MyData
# My Data now contain all numaric value 
```

```python
Tra = afterEnc.transform(MyDataCat)
Tra = Tra.toarray()
Tra
```

```python
ohe_df = pd.DataFrame(Tra, columns=afterEnc.get_feature_names())
```

```python
MyData.reset_index(drop=True, inplace=True)
ohe_df.reset_index(drop=True, inplace=True)
```

```python
#concat with original data
data = pd.concat([MyData, ohe_df], axis=1)
data
```

```python
Target = data.pop('price')
data
```

```python
%%time
from missingpy import MissForest 
# Make an instance and perform the imputation
imputer = MissForest()
X_imputed = imputer.fit_transform(data)
```

```python
# Add imputed values as columns to the untouched dataset
data['MF_kilometers'] = X_imputed[:, 3]
data['MF_owners'] = X_imputed[:, 4]
```

```python
# Now we Can Delete (Kilometers,owners) Columns , Since theres 2 alternative Column
data.drop("Kilometers",axis=1,inplace=True)
data.drop("owners",axis=1,inplace=True)
```

```python
data["engine_size"] = data["engine_size"].astype("int64")
data["MF_kilometers"] = data["MF_kilometers"].astype("int64")
data["MF_owners"] = data["MF_owners"].astype("int64")
data["passenger"] = data["passenger"].astype("int64")
```

```python
data
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, Target,train_size = 0.8, 
                                                    test_size = 0.2, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(bootstrap=True, ccp_alpha=0.0, criterion='mse', max_depth=60, max_features='auto', max_leaf_nodes=None,
                                  max_samples=None, min_impurity_decrease=0.0, min_impurity_split=None, min_samples_leaf=1,
                                  min_samples_split=2, min_weight_fraction_leaf=0.0, n_estimators=600, oob_score=False,
                                  random_state=42, verbose=2, warm_start=False,n_jobs=-1)
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

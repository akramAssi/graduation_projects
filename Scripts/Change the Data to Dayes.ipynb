```python
import pandas as pd
import os
```

```python
csv_Path = os.path.join("/Users/macbookpro/Desktop/For Graduation/Test","Data21.csv")
MyData = pd.read_csv(csv_Path,parse_dates = ['ad_date'])
MyData
```

```python
MyData["Days_Live"]=-999
```

```python
MyData.info()
```

```python
Count = 0
for row in range(0,MyData.shape[0]):
    date1 = date(MyData.at[row,"year"], 1, 1)
    date2 = MyData.at[row,"ad_date"].date()
    MyData.at[row,"Days_Live"] = (date2-date1).days
    Count = Count + 1
    
print("The Row is Affected was ---->",Count)    

    
```
